```python

dataX, dataY, xFeatures, yLabel = data_prep.feature_transform()
print('Full data Shape, dataX.shape = %s, dataY.shape = %s, len(xFeatures) = %s, yLabel = %s \n' % (
    str(dataX.shape), str(dataY.shape), str(len(xFeatures)), str(yLabel)))

trainX, trainY, testX, testY, cvalidX, cvalidY = data_prep.data_prep(dataX, dataY)
print('trainX.shape = %s, trainY.shape = %s, testX.shape = %s, testY.shape = %s, cvalidX.shape = %s, '
      'cvalidY.shape = %s \n'
      % (str(trainX.shape), str(trainY.shape), str(testX.shape), str(testY.shape), str(cvalidX.shape),
         str(cvalidY.shape)))
trainX, trainY = unison_shuffled_copies(trainX, trainY)

# Combine the test and validation data set
testX_nw = np.vstack((testX, cvalidX))
testY_nw = np.append(testY, cvalidY)
print ('testX_nw.shape = %s, testY_nw.shape = %s '%(str(testX_nw.shape), str(testY_nw.shape)))
```

# MODEL 1:
------------

The idea with Autoencoder lies in its core. We are trying to learn a representation by mapping the input to input. This representation can be thought as Latent feature set that is learned by the autoencoder module. Using the input we try to create a compresed representation of the data which lead to finding interesting patterns in the data. 

* The learning is accomplished by minimizing the Mean squared error between the input and the predicted input.
* 

### TRAIN AND CROSS VALIDATION

```python
tr_ls_arr, cv_ls_arr, tr_auc_arr, cv_auc_arr, tr_batch_mse, cv_batch_mse = autoencoders.train(
    trainX, trainY, cvalidX, cvalidY, layers=[15, 15], batch_size=32, display_step=1, num_epochs=100, 
    learning_rate=0.001, lamda=0.00001, REGULARIZE=False, MODEL_PATH=model_save_path, MODEL_NAME='autoencoder_model.ckpt'
)
print ('Mean Values: train_loss = %s, train_auc = %s, cv_loss = %s, cv_auc = %s'%(str(np.mean(tr_ls_arr)),  str(np.mean(cv_ls_arr)), 
                                                                                  str(np.mean(tr_auc_arr)), str(np.mean(cv_auc_arr))
                                                                                 ))
```

```python
oj = Plot(rows=1, columns=4, fig_size=(40,10))
tr_loss_df = pd.DataFrame(tr_ls_arr, columns=['training_loss'])
oj.vizualize(data=tr_loss_df, colX=None, colY=None, label_col=None, viz_type='line', params={'title':'Training Loss'})

tr_fpr, tr_tpr, threshold = Score.roc_curve(trainY, tr_batch_mse)
oj.vizualize(data=pd.DataFrame({'fpr':tr_fpr,'tpr':tr_tpr}), colX='fpr', colY='tpr', 
                 label_col=None, viz_type='roc',params={"title":"Train: ROC_CURVE"})

cv_loss_df = pd.DataFrame(cv_ls_arr, columns=['crossvalidation_loss'])
oj.vizualize(data=cv_loss_df, colX=None, colY=None, label_col=None, viz_type='line', params={'title':'Crossvalidation Loss'})

cv_fpr, cv_tpr, threshold = Score.roc_curve(cvalidY, cv_batch_mse)
oj.vizualize(data=pd.DataFrame({'fpr':cv_fpr,'tpr':cv_tpr}), colX='fpr', colY='tpr', 
                 label_col=None, viz_type='roc',params={"title":"Crossvalidation: ROC_CURVE"})
```

## CHECK IF THE FRAUD SCORE FOR FRADULENTS AND NON-FRADULENTS HAVE A DISTINCTION:
------------

It is quite obvious from the below plots that the Fraud-score for Non-fraudulent are majorly between 0 and 2, whereas the Fraud score for Fraudulents are very high and more that 90% of them are greater than 2.

```python
oj = Plot(1,2,fig_size=(40,12))
oj.vizualize(pd.DataFrame({'fraud_score':test_batch_mse[testY.flatten() == 0.0]}), colX='fraud_score', viz_type='hist',
        params={'title':'Fraud Score: For Non-Fraudulents', 'xlabel':'Fraud score', 'ylabel':'Count', 'bins':50})
oj.vizualize(pd.DataFrame({'fraud_score':test_batch_mse[(testY.flatten() == 0.0) & (test_batch_mse < 15)]}), 
             colX='fraud_score', viz_type='hist', params={'title':'Zoom IN: Fraud Score: For Nom-Fraudulents',
                                                          'xlabel':'Fraud score', 'ylabel':'Count', 'bins':50})
```

```python
oj = Plot(1,2,fig_size=(40,12))
oj.vizualize(pd.DataFrame({'fraud_score':test_batch_mse[testY.flatten() == 1.0]}), colX='fraud_score', viz_type='hist',
        params={'title':'Fraud Score: For Fraudulents', 'xlabel':'Fraud score', 'ylabel':'Count','bins':100})
oj.vizualize(pd.DataFrame({'fraud_score':test_batch_mse[(testY.flatten() == 1.0) & (test_batch_mse < 15)]}), 
             colX='fraud_score', viz_type='hist', params={'title':'Zoom IN: Fraud Score: For Fraudulents',
                                                          'xlabel':'Fraud score', 'ylabel':'Count','bins':100})
```
